```python
cols = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 
        'Distance', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt',
        'Regionname', 'Propertycount']
```

```python
df = df[cols]
df
```

```python
cols = ['Propertycount','Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols] = df[cols].fillna(0)
df
```

```python
df.isna().sum()
```

```python
df['Landsize'].mean()
```

```python

df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean()) 
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df.columns
```

```python
varlist =  ['Method_SA', 'Method_SP', 'Method_VB', 'Regionname_Eastern Victoria',
       'Regionname_Northern Metropolitan', 'Regionname_Northern Victoria',
       'Regionname_South-Eastern Metropolitan',
       'Regionname_Southern Metropolitan', 'Regionname_Western Metropolitan',
       'Regionname_Western Victoria']

# Defining the map function
def binary_map(x):
    return x.map({'True': 1, 'False': 0})

# Applying the function to the housing list
df[varlist] = df[varlist].astype(int)

df
```

```python
df.nunique()
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Price' , axis = 1)
y = df['Price']

X,y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape,y_train.shape
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_ , index = X.columns , columns = ['Coefficients/slopes'])
```

# Using Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=1.0,max_iter=None, tol=0.1,)
ridge.fit(X_train,y_train)
```

```python
ridge.score(X_train,y_train)
```

```python
ridge.score(X_test,y_test)
```

```python
ridge.coef_
```

```python
pd.DataFrame(ridge.coef_, index = X.columns , columns = ['Cofficients/slopes'])
```

# Using Lasso Regression 

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1.0,max_iter=1000,tol=0.0001)
lasso.fit(X_train,y_train)
```

```python
lasso.score(X_train,y_train)
```

```python
lasso.score(X_test,y_test)
```

```python
pd.DataFrame(lasso.coef_ , index = X.columns , columns = ['Coefficients/slopes'])
```
